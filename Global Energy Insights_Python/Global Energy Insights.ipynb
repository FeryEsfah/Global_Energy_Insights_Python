```python
def avg_gdp(merged_data_filter):
       
    merged_data_filter['avgGDP'] = merged_data_filter[[str(year) for year in range(2006, 2016)]].mean(axis=1)
    avg_gdp_sorted = merged_data_filter['avgGDP'].sort_values(ascending=False)
    
    return avg_gdp_sorted
    
avgGDP = avg_gdp(merged_data_filter)
avgGDP_df = pd.DataFrame(avgGDP, columns=['avgGDP'])
avgGDP_df
```

```python
def sixth_largest_gdp(final_data):
    avgGDP = final_data[['2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']].mean(axis=1)
    sixth_country = avgGDP.sort_values(ascending=False).index[5]
    gdp_change = final_data.loc[sixth_country, '2015'] - final_data.loc[sixth_country, '2006']
    return sixth_country, gdp_change

sixth_country, gdp_change = sixth_largest_gdp(final_data)
print("Country:", sixth_country)
print("GDP Change:", gdp_change)
```

```python
def energy_supply_capita(final_data):
    
    mean_energy_per_capita = final_data['Energy Supply per Capita'].mean()
    print("Mean energy supply per capita:", mean_energy_per_capita)
  
    final_data_clean = final_data.dropna(subset=['Energy Supply per Capita'])
    final_data_clean.loc[:, 'Above Mean'] = final_data_clean['Energy Supply per Capita'] > mean_energy_per_capita

    colors = []
    for value in final_data_clean['Above Mean']:
        if value:
            colors.append('purple')
        else:
            colors.append('yellow')

    plt.figure(figsize=(12, 6))
    plt.bar(final_data_clean.index, final_data_clean['Energy Supply per Capita'], color=colors)
    plt.axhline(mean_energy_per_capita, color='blue', linestyle='--', label='Mean Energy Supply per Capita')
    plt.xlabel('Countries')
    plt.ylabel('Energy Supply per Capita')
    plt.title('Energy Supply per Capita by Country')
    plt.xticks(rotation=80)
    plt.legend()
    plt.tight_layout()
    plt.show()

energy_supply_capita(final_data)
```

```python
#REPORT:

#Computed the mean energy supply per capita from the final_data DataFrame and printed the result, provided insight into
#the average energy supply levels across the dataset.

#Cleaned the data by removing rows with missing values in the 'Energy Supply per Capita' column, and added a new column 
#indicating whether each entry is above or below the mean value.

#Calculated and provided the minimum and maximum energy supply per capita values from the cleaned dataset, allowed for 
#quick identification of the range of energy supply levels among the countries.

#Added a new column, 'Above Mean',  to final_data_clean, indicating whether each country's energy supply per capita is 
#above the calculated mean. Used visualization different colors ('purple' for above mean and 'yellow' for below)
#to represent this distinction.

#Generated a bar chart  to display energy supply per capita by country, with a green dashed line indicating the mean value.
```

```python
def renewable_percentage(final_data):
    
    max_percentage = final_data['% Renewable'].max()
    min_percentage = final_data['% Renewable'].min()

    max_country = final_data[final_data['% Renewable'] == max_percentage].index[0]
    min_country = final_data[final_data['% Renewable'] == min_percentage].index[0]

    plt.figure(figsize=(12, 6))
    plt.bar(final_data.index, final_data['% Renewable'], color='skyblue', alpha=0.8, label='Countries')

    plt.bar(max_country, max_percentage, color='red', label='Max Renewable Country', edgecolor='black')
    plt.bar(min_country, min_percentage, color='orange', label='Min Renewable Country', edgecolor='black')
    plt.xlabel('Countries')
    plt.ylabel('% Renewable')
    plt.title('Renewable Energy Percentage by Country')
    plt.xticks(rotation=80)
    plt.axhline(max_percentage, color='green', linestyle='--', label='Maximum % Renewable')
    plt.axhline(min_percentage, color='purple', linestyle='--', label='Minimum % Renewable')
    plt.legend()
    plt.tight_layout()
    plt.show()

    print("Country with maximum % Renewable:", max_country)
    print("Maximum Percentage:", max_percentage)
    print("Country with minimum % Renewable:", min_country)
    print("Minimum Percentage:", min_percentage)

    return max_country, max_percentage, min_country, min_percentage
max_country, max_percentage, min_country, min_percentage = renewable_percentage(final_data)
```

```python
def self_citation_ratio(final_data):
    
    final_data.loc[:, 'Self-Citation Ratio'] = final_data['Self-citations'] / final_data['Citations']
    
    max_ratio = final_data['Self-Citation Ratio'].max()
    max_country = final_data[final_data['Self-Citation Ratio'] == max_ratio].index[0]
    
    return max_country, max_ratio

max_country, max_ratio = self_citation_ratio(final_data)

print("Country with highest Self-Citation Ratio:", max_country)
print("Ratio:", max_ratio)

plt.figure(figsize=(12, 6))
plt.bar(final_data.index, final_data['Self-Citation Ratio'], color='gray')
plt.axhline(max_ratio, color='red', linestyle='--', label=f'Max Self-Citation Ratio: {max_country} ({max_ratio:.2f})')
plt.xlabel('Countries')
plt.ylabel('Self-Citation Ratio')
plt.title('Self-Citation Ratio by Country')
plt.xticks(rotation=80)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def third_populous_country(final_data):
    final_data.loc[:, 'Estimated Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']
    sorted_data = final_data.sort_values(by='Estimated Population', ascending=False)
    third_country = sorted_data.index[2]  
    return third_country

third_country = third_populous_country(final_data)
print(f'Third most populous country according to the estimate: {third_country}')

final_data.loc[:, 'Estimated Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']

plt.figure(figsize=(12, 6))
plt.bar(final_data.index, final_data['Estimated Population'], color='blue', alpha=0.7)

third_population = final_data.loc[third_country, 'Estimated Population']
plt.axhline(third_population, color='red', linestyle='--', label=f'Third Most Populous Country: {third_country}')
plt.text(0.5, third_population + 1000000, f'{int(third_population):,}', color='black', ha='center')
plt.xlabel('Countries')
plt.ylabel('Estimated Population')
plt.title('Estimated Population by Country')
plt.xticks(rotation=80)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
def citable_correlation(final_data):
    
    final_data = final_data.dropna(subset=['Energy Supply', 'Energy Supply per Capita', 'Citable documents']).copy()
    final_data['Estimated Population'] = final_data['Energy Supply'] / final_data['Energy Supply per Capita']
    final_data['Citable Docs per Capita'] = final_data['Citable documents'] / final_data['Estimated Population']
    correlation = final_data['Citable Docs per Capita'].corr(final_data['Energy Supply per Capita'])
    
    plt.scatter(final_data['Energy Supply per Capita'], final_data['Citable Docs per Capita'], color='green')
    plt.title('Energy Supply per Capita vs. Citable Docs per Capita')
    plt.xlabel('Energy Supply per Capita')
    plt.ylabel('Citable Docs per Capita')
    plt.grid(False)
    plt.show()
    
    return correlation

correlation_value = citable_correlation(final_data)
print(f'Correlation: {correlation_value}')
```

```python
# REPORT:

#Calculated the correlation between "Energy Supply per Capita" and "Citable Documents per Capita" by estimating the
#population and dividing citable documents by population.

#Visualized the relationship between these two variables using a scatter plot, with energy supply on the x-axis and 
#citable documents on the y-axis.

#Printed calculated correlation value , indicated the strength and direction of the relationship between the two metrics.
```

```python
def high_renewable_indicator(final_data):

    sorted_data = final_data.sort_values(by='Rank')
    top_15 = sorted_data.head(15).copy()    
    median_renewable = top_15['% Renewable'].median()
    print("Median % Renewable for Top 15 Countries:", median_renewable)
    top_15.loc[:, 'HighRenew'] = (top_15['% Renewable'] >= median_renewable).astype(int)
    high_renew_series = top_15['HighRenew'].sort_index(ascending=True)
    
    return high_renew_series

high_renew_series = high_renewable_indicator(final_data)
print(high_renew_series)

plt.figure(figsize=(10, 6))
plt.bar(high_renew_series.index, high_renew_series, color=['gray' if val == 1 else 'red' for val in high_renew_series])
plt.xlabel('Countries')
plt.ylabel('High Renewable Indicator (1 = High, 0 = Low)')
plt.title('High Renewable Indicator for Top 15 Countries')
plt.xticks(rotation=80)
plt.yticks([0, 1], ['Low', 'High']) 
plt.axhline(0.5, color='blue', linestyle='--', label='Median Indicator')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# REPORT:

#Sorted countries by rank, selected the top 15, and calculated the median renewable energy percentage.

#Assigned a value of 1 if a country's renewable energy percentage is greater than or equal to the median and 0 otherwise,
#created a new column called HighRenew.

#Returned a series that indicates whether each country has a renewable energy percentage higher or lower than the median
#among the top 15 ranked countries.

#Visualized the number of countries with high versus low renewable energy by bar chart.
```

```python
energy = pd.read_excel('EnergyIndicators.xls', skiprows=17, usecols=[2, 3, 4, 5], nrows=227)
energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
energy.replace("...", pd.NA, inplace=True)
energy['Energy Supply'] = pd.to_numeric(energy['Energy Supply'], errors='coerce') * 1000000
energy['Country'] = energy['Country'].str.split('(').str[0].str.strip()
energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)
country_renames = {
    "Republic of Korea": "South Korea",
    "United States of America": "United States",
    "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
    "China, Hong Kong Special Administrative Region": "Hong Kong"
}
energy['Country'] = energy['Country'].replace(country_renames)

GDP = pd.read_csv('world_bank.csv', skiprows=4)
GDP.rename(columns={"Country Name": "Country"}, inplace=True)
GDP.replace({
    "Korea, Rep.": "South Korea",
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"
}, inplace=True)
GDP['Country'] = GDP['Country'].str.split('(').str[0].str.strip()
GDP['Country'] = GDP['Country'].str.replace(r'\d+', '', regex=True)

ScimEn = pd.read_excel('scimagojr-3.xlsx')
ScimEn['Country'] = ScimEn['Country'].str.split('(').str[0].str.strip()
ScimEn['Country'] = ScimEn['Country'].str.replace(r'\d+', '', regex=True)

merged_df = pd.merge(ScimEn, energy, how='inner', on='Country')
merged_df = pd.merge(merged_df, GDP, how='inner', on='Country')

merged_df['Estimated Population'] = merged_df['Energy Supply'] / merged_df['Energy Supply per Capita']

ContinentDict = {
    'China': 'Asia', 
    'United States': 'North America',
    'Japan': 'Asia', 
    'United Kingdom': 'Europe',
    'Russian Federation': 'Europe', 
    'Canada': 'North America', 
    'Germany': 'Europe', 
    'India': 'Asia', 
    'France': 'Europe', 
    'South Korea': 'Asia',
    'Italy': 'Europe', 
    'Spain': 'Europe', 
    'Iran': 'Asia', 
    'Australia': 'Australia', 
    'Brazil': 'South America'
}

merged_df['Continent'] = merged_df['Country'].map(ContinentDict)
valid_rows = merged_df.dropna(subset=['Energy Supply', 'Energy Supply per Capita']).copy()
valid_rows['Estimated Population'] = valid_rows['Energy Supply'] / valid_rows['Energy Supply per Capita']

continent_stats = valid_rows.groupby('Continent')['Estimated Population'].agg(['size', 'sum', 'mean', 'std'])

continent_stats
```

```python
def group_countries_renewable(merged_df, ContinentDict):

    merged_df['Continent'] = merged_df['Country'].map(ContinentDict)
    merged_df['% Renewable Bin'] = pd.cut(merged_df['% Renewable'], 5)
    grouped = merged_df.groupby(['Continent', '% Renewable Bin'], observed=False).size()

    return grouped

grouped = group_countries_renewable(merged_df, ContinentDict)
print(grouped)
```

# Q11) Visualization

```python
def visualize_grouped_data(grouped):
    
    grouped_unstacked = grouped.unstack()
    plt.figure(figsize=(10, 6))
    grouped_unstacked.plot(kind='bar', stacked=True, color=['pink','aqua','green','yellow','red'])

    plt.title('Number of Countries by Continent and % Renewable Bins', fontsize=14)
    plt.xlabel('Continent', fontsize=12)
    plt.ylabel('Number of Countries', fontsize=12)
    plt.xticks(rotation=80)
    plt.tight_layout()
    plt.legend(title='% Renewable Bin')
    plt.show()

visualize_grouped_data(grouped)
```

```python
# REPORT:

#Defined a dictionary to map countries to their respective continents and then added a new column to the DataFrame 
#(merged_df) that categorizes each country by continent using this mapping.

#Segmented the % Renewable data into five bins, created a new column (% Renewable Bin) that classifies countries based on
#their renewable energy percentage.

#Grouped the data by continent and the renewable energy bins, counted the number of countries in each group, resulted 
#in a structured overview of renewable energy distribution across continents.

#Converted this grouped data into a DataFrame (grouped_df) for visualization.

#Created a bar chart to display the number of countries per continent across the renewable energy bins. Represented each continent 
#in the chart, allowed for an easy comparison of renewable energy distribution among different continents. Clarified the chart with adding labeled axes,
#a title, and a legend.
```

```python
def population_series(merged_data_filter):

    merged_data_filter = merged_data.nsmallest(15, 'Rank')
    merged_data_filter['Estimated Population'] = merged_data_filter['Energy Supply'] / merged_data_filter['Energy Supply per Capita']
    merged_data_filter['Estimated Population'] = pd.to_numeric(merged_data_filter['Estimated Population'], errors='coerce')
    merged_data_filter['Estimated Population'] = merged_data_filter['Estimated Population'].map(lambda x: "{:,.8f}".format(x) if pd.notnull(x) else "NaN")
    PopEst = pd.DataFrame(merged_data_filter['Estimated Population'])
    PopEst.index = merged_data_filter['Country']

    return PopEst

PopEst = population_series(merged_data_filter)
PopEst
```

# Q12) Visualization

```python
def population_estimate(PopEst):
    
    if PopEst['Estimated Population'].dtype == 'object':
        PopEst['Estimated Population'] = PopEst['Estimated Population'].str.replace(",", "").astype(float)

    plt.figure(figsize=(12, 6))
    plt.bar(PopEst.index, PopEst['Estimated Population'], color=['aqua'], alpha=0.7)
    plt.title('Estimated Population of the 15 Countries with Lowest Rank')
    plt.xlabel('Country')
    plt.ylabel('Estimated Population')
    plt.xticks(rotation=80)
    plt.tight_layout()
    plt.show()

population_estimate(PopEst)
```
